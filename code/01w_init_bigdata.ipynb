```python
import pandas as pd
pop=pd.read_csv("popmonth.csv",encoding='cp949',index_col = 0, thousands = ',')
pop
```

```python
pop.columns= ['총인구수','세대수','세대당인구','남자인구수','여자인구수','남여비율']
pop.head()
```

```python
pop.index[0].split()[0]
```

```python
for i in range(len(pop)):
    print(pop.index[i])
```

```python
pop.index = [pop.index[i].split()[0] for i in range(len(pop))]
pop
```

```python
type(pop)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (5,3))
sns.scatterplot(pop['총인구수'])
plt.xticks(rotation = 90)
plt.show()
```

```python
pop.sort_values('총인구수')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
pop1=pop.sort_values('총인구수')
sns.barplot(x=pop1.index,y='총인구수',data =pop1)
plt.xticks(rotation = 90)
plt.show()
```
