```python
# df.query(조건) - 조건을 만족하는 행 표시
df = pd.DataFrame({'A': range(1, 6),
                   'B': range(6, -3, -2),
                   'C': range(10, 5, -1),
                   'D clmns': range(3, -2, -1),
                   'E': ['python', 'java', 'python', 'go', 'python']})
df
```

```python
df.query('A > B')
```

```python
df[df.A > df.B]
```

```python
df.query('C == 10')
```

```python
df.query('A > 0')
```

```python
df.query('E == "python"')
```

```python
df.query("E != 'python'")
```

```python
# 열명에 공백이 있는 경우 → 백쿼트(`) 사용
df.query('A == `D clmns`')               # A열 == D clmns열
```

```python
df.query("A == 'D clmns'")               # 이건 문자열 비교 -> 결과 없음
```

```python
df.query('A > 3 | B > 8')
```

```python
df.query('A > 2 & C < 8')
```

```python
df.query("E in ['python', 'go']")
```

```python
num = 3
df.query('B > @num')  # @변수명
```

```python
num = 3
df.query(f'B > {num}')   # f-string
```

```python
s = "E.str.contains('py')"  # E열에 py 포함
df.query(s)
```

```python
s = 'E.str.startswith("g")'
df.query(s)                               # E열이 'g'로 시작
```
